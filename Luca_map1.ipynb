```python
import matplotlib.pyplot as plt
import numpy as np

parteien = ['SVP', 'SP', 'Mitte', 'FDP', 'Grüne', 'GLP', 'MCG', 'EVP', 'EDU', 'Lega', 'PdA/Sol.']
sitze = [62, 41, 29, 28, 23, 10, 2, 2, 2, 1, 0]
farben = ['#003366', '#FF0000', '#FFCC00', '#0066CC', '#009933', '#66CC66', '#CC00FF', '#FF9900', '#660099', '#000000', '#FF3399']
aenderungen = [9, 2, 1, -1, 5, -6, 2, -1, 1, 0, -2]
manuelle_reihenfolge = ['SVP', 'SP', 'Mitte', 'FDP', 'Grüne', 'GLP', 'MCG', 'EVP', 'EDU', 'Lega', 'PdA/Sol.']
sortierte_daten = sorted(zip(parteien, sitze, farben, aenderungen), key=lambda x: manuelle_reihenfolge.index(x[0]))
sortierte_parteien, sortierte_sitze, sortierte_farben, sortierte_aenderungen = zip(*sortierte_daten)
umgekehrte_sitze = list(reversed(sortierte_sitze))
umgekehrte_farben = list(reversed(sortierte_farben))
umgekehrte_aenderungen = list(reversed(sortierte_aenderungen))
umgekehrte_parteien = list(reversed(sortierte_parteien))

fig, ax = plt.subplots(figsize=(15, 8))
bar_positions = np.arange(len(umgekehrte_parteien))
bars = ax.barh(bar_positions, umgekehrte_sitze, color=umgekehrte_farben, edgecolor='black', height=0.4)

ax.text(-15, len(umgekehrte_parteien) - 0.5, 'Partei', ha='right', va='center', fontsize=14, fontweight='bold')
ax.text(-5, len(umgekehrte_parteien) - 0.5, 'Sitze', ha='right', va='center', fontsize=14, fontweight='bold')
ax.text(sitze[0] + 10, len(umgekehrte_parteien) - 0.5, '+/-', ha='center', va='center', fontsize=14, fontweight='bold')

for y, partei in zip(bar_positions, umgekehrte_parteien):
    ax.text(-15, y, partei, ha='right', va='center', fontsize=12)
for y, sitz in zip(bar_positions, umgekehrte_sitze):
    ax.text(-5, y, sitz, ha='right', va='center', fontsize=12)
for y, change in zip(bar_positions, umgekehrte_aenderungen):
    if change > 0:
        ax.text(sitz + 5, y, '↑', ha='center', va='center', fontsize=20, color='green')
    elif change < 0:
        ax.text(sitz + 5, y, '↓', ha='center', va='center', fontsize=20, color='red')
    else:
        ax.text(sitz + 5, y, '•', ha='center', va='center', fontsize=20, color='grey')
for y, change in zip(bar_positions, umgekehrte_aenderungen):
    ax.text(sitz + 10, y, f'{change:+}', ha='center', va='center', fontsize=12)

ax.axis('off')
plt.title('Schweiz Endresultat Nationalrat', fontsize=20, fontweight='bold')
plt.tight_layout()
plt.show()
```
